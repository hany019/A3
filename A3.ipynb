```python
# Loading the libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from minisom import MiniSom

# Define colors for different classes in the synthetic dataset
colors_synthetic = {
    1: 'red',
    2: 'blue',
    3: 'green',
    4: 'yellow',
    5: 'orange',
    6: 'purple'
}

# Loading the Synthetic dataset
data_path_synthetic = 'A3-data.csv'
synthetic_data = pd.read_csv(data_path_synthetic)
print(synthetic_data.columns)

# Adjusting colors_synthetic to match the number of observations minus one for Synthetic Data
colors_synthetic = synthetic_data['class'].apply(lambda x: colors_synthetic[x]).tolist()

# Loading the Glass dataset
data_path_glass = 'glass.csv'
glass_data = pd.read_csv(data_path_glass)

# Define colors for different types in the glass dataset
colors_glass = {
    1: 'red',
    2: 'blue',
    3: 'green',
    4: 'yellow',
    5: 'orange',
    6: 'purple',
    7: 'cyan'
}

# Adjusting colors_glass to match the number of observations minus one for Glass Data
colors_glass = glass_data['Type'].apply(lambda x: colors_glass[x]).tolist()

# For synthetic dataset - Extracting features and excluding the class information ('class' column)
synthetic_features = synthetic_data.drop('class', axis=1)

# For glass dataset 
glass_features = glass_data.drop('Type', axis=1)

# Standardize the data
scaler = StandardScaler()

# Synthetic Data
synthetic_features_scaled = scaler.fit_transform(synthetic_features)

# Glass Data
glass_features_scaled = scaler.fit_transform(glass_features)

# PCA
# Applying PCA
pca = PCA(n_components=2)
synthetic_pca = pca.fit_transform(synthetic_features_scaled)
glass_pca = pca.fit_transform(glass_features_scaled)

# PCA plot for Synthetic Data
plt.figure(figsize=(6, 6))
plt.scatter(synthetic_pca[:, 0], synthetic_pca[:, 1], cmap='viridis', c=synthetic_data['class'])
plt.colorbar(label='Class')
plt.title('PCA - Synthetic Data')
plt.savefig('PCA_Synthetic.png')  # Save PCA Plot for Synthetic Data

# PCA plot for Glass Data
plt.figure(figsize=(6, 6))
plt.scatter(glass_pca[:, 0], glass_pca[:, 1], cmap='viridis', c=glass_data['Type'])
plt.colorbar(label='Type')
plt.title('PCA - Glass Data')
plt.savefig('PCA_Glass.png')  # Save PCA Plot for Glass Data

# Scree plot for synthetic data
plt.figure(figsize=(6, 4))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components (Synthetic Data)')
plt.grid()
plt.show()

# Apply PCA for glass data
pca_glass = PCA().fit(glass_features_scaled)

# Scree plot for glass data
plt.figure(figsize=(6, 4))
plt.plot(np.cumsum(pca_glass.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components (Glass Data)')
plt.grid()
plt.show()

# t-SNE
# Applying t-SNE for Synthetic Data
tsne_params = [5, 30]  # Example perplexity values, We test this Hany
for i, perplexity in enumerate(tsne_params):
    tsne = TSNE(n_components=2, perplexity=perplexity)
    synthetic_tsne = tsne.fit_transform(synthetic_features_scaled)
    
    plt.figure(figsize=(6, 6))
    plt.scatter(synthetic_tsne[:, 0], synthetic_tsne[:, 1], cmap='viridis', c=synthetic_data['class'])
    plt.colorbar(label='Class')
    plt.title(f't-SNE (Perplexity={perplexity}) - Synthetic Data')
    plt.savefig(f'tSNE_Perplexity_{perplexity}_Synthetic.png')  # Save t-SNE Plot for Synthetic Data with different perplexity values

# Applying t-SNE for Glass Data
tsne_params_glass = [5, 30]  # Example perplexity values for Glass Data, We test this Hany
for i, perplexity in enumerate(tsne_params_glass):
    tsne_glass = TSNE(n_components=2, perplexity=perplexity)
    glass_tsne = tsne_glass.fit_transform(glass_features_scaled)
    
    plt.figure(figsize=(6, 6))
    plt.scatter(glass_tsne[:, 0], glass_tsne[:, 1], cmap='viridis', c=glass_data['Type'])
    plt.colorbar(label='Type')
    plt.title(f't-SNE (Perplexity={perplexity}) - Glass Data')
    plt.savefig(f'tSNE_Perplexity_{perplexity}_Glass.png')  # Save t-SNE Plot for Glass Data with different perplexity values

# K-means
# Getting the unique values in 'class' column for synthetic data and 'Type' column for glass data
unique_classes_synthetic = synthetic_data['class'].unique()
unique_classes_glass = glass_data['Type'].unique()

# Applying K-Means clustering and comparing with real classes for Synthetic Data
real_classes_synthetic = synthetic_data['class'] 
real_number_of_classes_synthetic = len(unique_classes_synthetic)
real_number_of_classes_glass = len(unique_classes_glass)

# For Synthetic data
real_classes_synthetic = synthetic_data['class']

for k in range(2, 6):
    kmeans_synthetic = KMeans(n_clusters=k, random_state=42)
    synthetic_kmeans = kmeans_synthetic.fit_predict(synthetic_features_scaled)

    # Compare obtained cluster labels with real labels for each iteration of K
    cm = confusion_matrix(real_classes_synthetic, synthetic_kmeans)
    print(f"Confusion Matrix for K = {k}:")
    print(cm)

    # K-Means plot for Synthetic Data
    plt.figure(figsize=(6, 6))
    plt.scatter(synthetic_pca[:, 0], synthetic_pca[:, 1], c=synthetic_kmeans, cmap='viridis')
    plt.colorbar(label='Cluster')
    plt.title(f'KMeans Clustering (K={k}) - Synthetic Data')
    plt.savefig(f'KMeans_Synthetic_K_{k}.png')  # Save KMeans Plot for Synthetic Data
    plt.show()
    
    # Evaluate for  equals the real number of classes
    if k == real_number_of_classes_synthetic:
        print(f"Confusion Matrix for K = {real_number_of_classes_synthetic} (Real number of classes):")
        print(cm)
    
# For Glass data
real_classes_glass = glass_data['Type']

for k in range(2, 6):
    kmeans_glass = KMeans(n_clusters=k, random_state=42)
    glass_kmeans = kmeans_glass.fit_predict(glass_features_scaled)

    # Compare obtained cluster labels with real labels for each iteration of K
    cm_glass = confusion_matrix(real_classes_glass, glass_kmeans)
    print(f"Confusion Matrix for K = {k}:")
    print(cm_glass)

    # K-Means plot for Glass Data
    plt.figure(figsize=(6, 6))
    plt.scatter(glass_pca[:, 0], glass_pca[:, 1], c=glass_kmeans, cmap='viridis')
    plt.colorbar(label='Cluster')
    plt.title(f'KMeans Clustering (K={k}) - Glass Data')
    plt.savefig(f'KMeans_Glass_K_{k}.png')  # Save KMeans Plot for Glass Data
    plt.show()
    
    # Evaluate for  equals the real number of classes
    if k == real_number_of_classes_glass:
        print(f"Confusion Matrix for K = {real_number_of_classes_glass} (Real number of classes):")
        print(cm_glass)

# AHC
from scipy.cluster.hierarchy import dendrogram, linkage

# Synthetic Dataset - Applying AHC with UPGMA for Synthetic Data
ahc_synthetic_upgma = AgglomerativeClustering(n_clusters=4, linkage='average')  # UPGMA
synthetic_ahc_upgma = ahc_synthetic_upgma.fit(synthetic_features_scaled)

# Assigning different colors to each original class in Synthetic Data
unique_labels_synthetic = synthetic_data['class'].unique()
colors_synthetic = {label: idx for idx, label in enumerate(unique_labels_synthetic)}
colors_synthetic = [colors_synthetic[label] for label in synthetic_data['class']]

# Adjusting colors_synthetic to match the number of observations minus one for Synthetic Data
colors_synthetic = synthetic_data['class'].apply(lambda x: colors_synthetic[x]).tolist()

# Make sure the length of colors_synthetic matches the number of observations minus one
if len(colors_synthetic) != len(synthetic_data) - 1:
    colors_synthetic = synthetic_data['class'].apply(lambda x: colors_synthetic[x]).tolist()[:-1]

# Creating the linkage matrix required for dendrogram with UPGMA
linkage_matrix_synthetic_upgma = linkage(ahc_synthetic_upgma.children_, method='average')

# AHC Dendrogram plot for Synthetic Data with UPGMA (with different colors for each class)
plt.figure(figsize=(10, 5))
plt.title('Dendrogram - Synthetic Data (UPGMA)')
dendrogram(linkage_matrix_synthetic_upgma, labels=colors_synthetic)  # Using labels parameter for color coding
plt.savefig('Dendrogram_Synthetic_UPGMA_ColorCoded.png')  # Save AHC Dendrogram Plot for Synthetic Data with UPGMA

# Glass Dataset - Applying AHC with complete linkage for Glass Data
ahc_glass_complete = AgglomerativeClustering(n_clusters=4, linkage='complete')  # Complete linkage
glass_ahc_complete = ahc_glass_complete.fit(glass_features_scaled)

# Adjusting colors_glass to match the number of observations for Glass Data
colors_glass = glass_data['Type'].apply(lambda x: colors_glass[x]).tolist()

# Creating the linkage matrix required for dendrogram with complete linkage for Glass Data
linkage_matrix_glass_complete = linkage(glass_ahc_complete.children_, method='complete')

# Get the number of observations from the linkage matrix
num_observations = len(linkage_matrix_glass_complete) + 1

# Truncate colors_glass to match the number of observations
colors_truncated = colors_glass[:num_observations]

# AHC Dendrogram plot for Glass Data with complete linkage (with different colors for each class)
plt.figure(figsize=(10, 5))
plt.title('Dendrogram - Glass Data (Complete Linkage)')
dendrogram(linkage_matrix_glass_complete, labels=colors_truncated)  # Use truncated labels matching the number of observations
plt.savefig('Dendrogram_Glass_Complete_ColorCoded.png')  # Save AHC Dendrogram Plot for Glass Data with complete linkage

# SOM
# SOM visualization for Synthetic Data
som_synthetic = MiniSom(10, 10, synthetic_features_scaled.shape[1], sigma=0.5, learning_rate=0.5)
som_synthetic.random_weights_init(synthetic_features_scaled)
som_synthetic.train_random(synthetic_features_scaled, 100)

# SOM Heatmap plot for Synthetic Data
plt.figure(figsize=(6, 6))
plt.title('SOM Heatmap - Synthetic Data')
plt.pcolor(som_synthetic.distance_map().T, cmap='bone_r')  # plotting the distance map as a heatmap
plt.colorbar()
plt.savefig('SOM_Heatmap_Synthetic.png')  # Save SOM Heatmap for Synthetic Data

# U-Matrix Visualization for Synthetic Data
umatrix_synthetic = som_synthetic.distance_map()
plt.figure(figsize=(6, 6))
plt.title('U-Matrix - Synthetic Data')
plt.imshow(umatrix_synthetic, cmap='gray')  # U-Matrix visualization
plt.colorbar()
plt.savefig('SOM_Umatrix_Synthetic.png')  # Save U-Matrix for Synthetic Data

# Component Planes Visualization for Synthetic Data (assuming 2D SOM)
plt.figure(figsize=(10, 10))
for i, f in enumerate(synthetic_features.columns):
    plt.subplot(4, 4, i+1)
    plt.title(f'Component Plane - {f}')
    plt.imshow(som_synthetic.get_weights()[:, :, i], cmap='coolwarm')
    plt.colorbar()
plt.tight_layout()
plt.savefig('SOM_ComponentPlanes_Synthetic.png')  # Save Component Planes for Synthetic Data

# SOM visualization for Glass Data
som_glass = MiniSom(10, 10, glass_features_scaled.shape[1], sigma=0.5, learning_rate=0.5)
som_glass.random_weights_init(glass_features_scaled)
som_glass.train_random(glass_features_scaled, 100)

# SOM Heatmap plot for Glass Data
plt.figure(figsize=(6, 6))
plt.title('SOM Heatmap - Glass Data')
plt.pcolor(som_glass.distance_map().T, cmap='bone_r')  # plotting the distance map as a heatmap
plt.colorbar()
plt.savefig('SOM_Heatmap_Glass.png')  # Save SOM Heatmap for Glass Data

# U-Matrix Visualization for Glass Data
umatrix_glass = som_glass.distance_map()
plt.figure(figsize=(6, 6))
plt.title('U-Matrix - Glass Data')
plt.imshow(umatrix_glass, cmap='gray')  # U-Matrix visualization
plt.colorbar()
plt.savefig('SOM_Umatrix_Glass.png')  # Save U-Matrix for Glass Data

# Component Planes Visualization for Glass Data (assuming 2D SOM)
num_features = len(glass_features.columns)
num_rows = (num_features + 1) // 4 + 1  # Calculate the number of rows needed for subplots
plt.figure(figsize=(15, 5 * num_rows))  # Adjusting the figure size based on the number of rows needed

for i, f in enumerate(glass_features.columns):
    plt.subplot(num_rows, 4, i+1)
    plt.title(f'Component Plane - {f}')
    plt.imshow(som_glass.get_weights()[:, :, i], cmap='coolwarm')
    plt.colorbar()

plt.tight_layout()
plt.savefig('SOM_ComponentPlanes_Glass.png')  # Save Component Planes for Glass Data
```
